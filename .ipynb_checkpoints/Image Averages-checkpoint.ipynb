Lets go ahead and get the average for all the different classes of images.

```python
categories = ['airplane',
 'automobile',
 'bird',
 'cat',
 'deer',
 'dog',
 'frog',
 'horse',
 'ship',
 'truck']
for i, category in enumerate(categories):
    imgs_in_category = training_data[labels['label'] == category]
    avg = imgs_in_category.mean(axis=0)
    plt.figure(i)
    plt.imshow(avg, cmap='Greys')
    plt.title(category)
    plt.draw()
```

Heres a quick reminder for the first challenge.


"Challenge 1: Use the average image to compute the class of an image
Since images from each class look slightly different, it might be worthwhile to use that to predict the classes of our images. The code snippets provided might help you in this.
- Step 1: Compute the average image for each class. Much of this has been done for you.
- Step 2: For each image in the test set, compute the sum of squared difference with the pixels of each "averaged" image. You will obtain 10 quantities. Assign the label to this test image which corresponds with the smallest difference.
- Step 3: Put your results in a csv, upload it to kaggle, and add a screenshot to a folder with your code. Upload all this to github. Remember: you do NOT need to upload the data to github."

```python
#compute the average image for each class
avg_list = []
for i, category in enumerate(categories):
    imgs_in_category = training_data[labels['label'] == category]
    avg_list.append(imgs_in_category.mean(axis=0))
```

```python
# load the test set (this will take about one minute)
test_ = []
for i in np.arange(1, 300001):
    test_.append(misc.imread('/media/a-stahl/Windows/Users/Alex/Documents/Fall 2017/test/test/' + str(i) + '.png', mode='I'))
```

```python
# Convert the list into a panel
testing_data = pd.Panel(test_)
testing_data
```

Below is our model for comparing the test image with all ten average category images and choosing the most similar category as the label.

```python
def averageComparisonModel(num):
    """Subtracts current image from all 10 possible image averages and returns the one with the lowest total
        returns the estimated image"""
    avg_compared = [] #list holding current img - each 10 averages
    for j in range(len(avg_list)):
        sumSqrd = ((testing_data.iloc[num - 1] - avg_list[j]) ** 2).sum().sum()
        avg_compared.append(sumSqrd) #compute sum of squared differences and create list of all labels
        minCost = min(avg_compared) # select the smallest difference
    return avg_compared.index(minCost)
    
```

```python
#Lets look at a few examples to make sure our model is working
testval = averageComparisonModel(0)
print('Prediction = ' + categories[testval])

#Now lets display the image
sample_img_color = misc.imread('/media/a-stahl/Windows/Users/Alex/Documents/Fall 2017/test/test/5.png', mode='RGB') 
plt.imshow(sample_img_color)
plt.axis('off')
```

```python
#Once more
testval = averageComparisonModel(234235)
print('Prediction = ' + categories[testval])
sample_img_color = misc.imread('/media/a-stahl/Windows/Users/Alex/Documents/Fall 2017/test/test/234235.png', mode='RGB') 
plt.imshow(sample_img_color)
plt.axis('off')
```

Good! Looks like our output is in the correct format, even though our guess isnt always correct. I'm assuming it predicted automobile for this one because the bottom half looks like a road.

Lets run it for all 300,000 test images. This will take some time depending on the hardware. (around 30 min on my laptop)

```python
#this may take a while
solution = [(i, categories[averageComparisonModel(i)]) for i in np.arange(1,300001)]
```

Finally, lets write the results to a csv file so we can upload it.

```python
with open('solution.csv', 'w') as f:
    writer = csv.writer(f)
    writer.writerow(('id', 'label'))
    writer.writerows(solution)
```
